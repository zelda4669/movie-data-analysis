# Domestic Profit Analysis

## Overview

This project analyzes box office and streaming data to provide strategic direction for a newly formed movie division. Our analysis explores financial data, film ratings, and viewership data to help guide the division in deciding what types of movies to create.

## Notebook Overview

Critical for any new business venture is to add value as quickly as possible. This notebook will analyze financial information for movies released from 2010-2019 to determine what types of movies produce the greatest returns.

```python
#import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
%matplotlib inline

#import datasets
domestic_profit = pd.read_csv('data/Edited Data/Domestic Profits.csv')
movie_details = pd.read_csv('data/Provided Data/tmdb.movies.csv', index_col = 0, converters={'genre_ids': eval})
genres = pd.read_csv('data/edited data/tmdb genres.csv')
```

```python
domestic_profit.head()
```

```python
domestic_profit.describe()
```

A quick look at the summary statistics for this data shows that the median "profit" for movies is a loss nearly $1.2 million. This highlights the importance of proceeding thoughtfully--it is just as easy to lose money as it is to make money. Let's take a look at the movies with the highest ROI to see if there are any similarities.

```python
ROI = domestic_profit.sort_values(by='domestic_ROI', ascending=False).head(50)
```

```python
ROI.describe()
```

The films with the highest ROI have a mean budget of \\$4.3 million (significantly less than the mean of \\$36 million for the entire dataset) and a median budget of \\$2.8 million (compared to a median budget of \\$16 million for the entire dataset), indicating that modest budgets provide a greater return then big budget movies.

```python
#graph budgets compared to ROI as a percentage
fig = plt.figure(figsize=(12,10))
plt.ylim(-200, 1600)
plt.xlim(-10000000, 280000000)
ax1 = sns.scatterplot(data=domestic_profit, x='production_budget', y='domestic_ROI')
ax1.set(xlabel='Budget (1 = 100 million)', ylabel='ROI (Percentage)', title='Return on Investment by Budget Amount');
```

Graphing all of our movie budgets gives us a similar picture! For movies with budgets over \\$100 million, ROI caps out around 250\%, but for movies with budgets less than \\$500,000, there is greater potential to make a larger return.

Let's look at the movies with the highest rate of return to see if we can find any similarities between them.

```python
top_roi_details = ROI.merge(movie_details, how='inner', left_on='movie', right_on='title')
top_roi_details.drop_duplicates(subset=['movie'], inplace=True)
```

```python
#define a function that counts the frequency of genres in the genre_id column
def genre_count(series):
    g = []
    genres = {}
    for i in series:
        g.extend(i)
    x = list(set(g))
    for i in x:
        genres[i] = 0
    for i in g:
        genres[i] += 1
    df = pd.DataFrame.from_dict(genres, orient='index')
    df.sort_values(by=0, inplace=True, ascending=False)
    return df
```

```python
#use the function to create a bar graph of genres that occur most frequently
chart = genre_count(top_roi_details['genre_ids'])
chart = chart.merge(genres, left_index=True, right_on='id')
fig = plt.figure(figsize=(15,8))
ax2 = sns.barplot(data=chart, y=0, x='name', palette = 'Purples_r')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=40, ha="right")
ax2.set(xlabel='Genre', ylabel='Count', title='Genres With Best ROI');
```

Horror, drama, and thriller outperform every other genre in this subset of data! Let's look at the genre counts as a whole to make sure this can't be attributed to the distribution of genres in the full dataset.

```python
chart = genre_count(movie_details['genre_ids'])
chart = chart.merge(genres, left_index=True, right_on='id')
fig = plt.figure(figsize=(15,8))
ax3 = sns.barplot(data=chart, y=0, x='name', palette = 'Purples_r')
ax3.set_xticklabels(ax3.get_xticklabels(), rotation=40, ha="right")
ax3.set(xlabel='Genre', ylabel='Count', title='Total Genre Distribution');
```

Drama is by far the most popular genre for movies in the entire dataset, making its appearance at the top of our subset graph less impactful. Horror and thriller, on the other hand, are clearly over-represented in our data subset, so we can be more confident that those genres provide higher ROI.

## Conclusions

Based on this data, my recommendations for Microsoft are:

- **Big budget does not equal big returns**. Movies with a budget less than \\$500,000 provide opportunity for the highest rate of return.
- **Horror and thriller movies are disproportionately present amongst movies with the highest rate of return**. Drama is also a heavily represented genre, but it is also the most represented genre overall so it is challenging to determine how meaningful that is without further analysis.
